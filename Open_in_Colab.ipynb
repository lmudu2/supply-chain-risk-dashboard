# Supply Chain Risk Dashboard - Native Colab Deployment

This dashboard is optimized for Google Colab using native proxying and automated setup.

### 🚀 Instructions
1. Click **Runtime > Run All**.
2. Wait for the repo to clone and the dashboard to start.
3. Click the **Dashboard Ready** link that appears at the bottom.

```python
# --- 1. SETUP & REPO CLONING ---
import os
import time
import socket
from google.colab import output

# Cleanup old sessions
!fuser -k 8501/tcp
!rm -rf supply-chain-risk-dashboard
!git clone https://github.com/lmudu2/supply-chain-risk-dashboard.git
%cd supply-chain-risk-dashboard
!pip install -r requirements.txt --quiet

# --- 2. START DASHBOARD ---
def wait_for_port(port, timeout=30):
    start_time = time.time()
    while time.time() - start_time < timeout:
        with socket.socket(socket.AF_INET, socket.SOCK_STREAM) as s:
            if s.connect_ex(("localhost", port)) == 0: return True
        time.sleep(1)
    return False

# Launch Streamlit
get_ipython().system_raw("streamlit run app.py --server.port 8501 --server.enableCORS False --server.headless True &")

if wait_for_port(8501):
    print("\n🚀 DASHBOARD READY!")
    print("Click the link below to open:")
    display(output.eval_js(f"google.colab.kernel.proxyPort(8501)"))
else:
    print("\n❌ Error: Dashboard failed to start.")
```
